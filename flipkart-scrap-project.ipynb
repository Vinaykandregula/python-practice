```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
flipkart_link='https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=25a6a97e-767c-4032-bb1e-d2f9b2ae9823&sort=popularity'
```

```python
import os
def wrapping(flipkart_link):
    response=requests.get(flipkart_link)
    if response.status_code!=200:
        raise Exception('unable to load link')
        return
    info=BeautifulSoup(response.content,'lxml')
    mobile_tag=info.find_all('div',class_='_4rR01T')
    rating_tag=info.find_all('span',class_='_2_R_DZ')
    price_tag=info.find_all('div',class_='_30jeq3 _1_WHN1')
    def mobile_details(mobile_tag,rating_tag,price_tag):
        mobile=mobile_tag.text.strip()
        rating=rating_tag.text.split()[0]
        price=price_tag.text.strip()
        return mobile,rating,price
    pop={'Mobile':[],'Rating':[],'Price':[]}
    for i in range(len(mobile_tag)):
        Details=mobile_details(mobile_tag[i],rating_tag[i],price_tag[i])
        pop['Mobile'].append(Details[0])
        pop['Rating'].append(Details[1])
        pop['Price'].append(Details[2])
    return pd.DataFrame(pop)
df=wrapping(flipkart_link)
df.to_csv('flipkart popularity mobiles.csv')
```
