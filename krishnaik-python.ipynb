### Asynchronous Execution(Parallel Execution) :

```python
import time
from concurrent.futures import ThreadPoolExecutor
def num(a):
    time.sleep(1)
    return a
```

```python
start=time.time()
with ThreadPoolExecutor(max_workers=1) as executor:   #here executor takes threads as list.each thread is one element in list
    for result in executor.map(num,range(10)):
        print('count: {0}'.format(result))
print('The total time: {0}'.format(time.time()-start))
```
